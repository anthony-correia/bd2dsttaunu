# MC sample of $B \to D^{*-}\left(D^{+} \to K^- \pi^+ \pi^+  \right)X$ Background

A fit to the $m(K\pi\pi)$ invariant mass in `MC` is performed in order to get the shape of the $D^+$ peak.\
This is used to fix the tail parameters of the `data` sample.

**Model**: The $D^+$ peak in $m(K\pi\pi)$ is modeled by the sum of two CB PDFs with tails in opposite sides and shared $\mu$ and $\sigma$

```python
obs = zfit.Space("x", limits=(low_MC, high_MC))
num += 1
zparams = HEA.fit.fit.define_zparams(initial_values, num=num)
```

### PDFs

```python
# doubleCB = zfit.pdf.DoubleCB(zparams['mu'], zparams['sigma'], 
#                              zparams['alphaL'], zparams['alphaR'], 
#                              zparams['nL'], zparams['nR'], obs=obs)

from HEA.fit.fit import sum_crystalball_or_gaussian
## Shared mu and sigma
model_S, crystalBallL_S, crystalBallR_S = sum_crystalball_or_gaussian(zparams['mu'], zparams['mu'],
                                                                      zparams['sigma'], zparams['sigma'],
                                                                      zparams['frac'], obs,                                                          
                                                                      zparams['alphaL'], zparams['alphaR'], 
                                                                      zparams['nL'], zparams['nR']
                                                                     )

## Shared mu and different sigma
# model_S, crystalBallL_S, crystalBallR_S = sum_crystalball_or_gaussian(zparams['mu'], zparams['mu'],
#                                                                       zparams['sigmaL'], zparams['sigmaR'],
#                                                                       zparams['frac'], obs,                                                          
#                                                                       zparams['alphaL'], zparams['alphaR'], 
#                                                                       zparams['nL'], zparams['nR']
#                                                                      )

# # Different mu and sigma
# model_S, crystalBallL_S, crystalBallR_S = sum_crystalball_or_gaussian(zparams['muL'], zparams['muR'],
#                                                                       zparams['sigmaL'], zparams['sigmaR'],
#                                                                       zparams['frac'], obs,                                                          
#                                                                       zparams['alphaL'], zparams['alphaR'], 
#                                                                       zparams['nL'], zparams['nR']
#                                                                      )

data = zfit.Data.from_pandas(df[branch], obs=obs)
```

### fit

```python
# result, params = HEA.fit.fit.launch_fit(doubleCB, data, extended=False)
# result, params = HEA.fit.fit.launch_fit(doubleCB, data, extended=False)
result, params = HEA.fit.fit.launch_fit(model_S, data, extended=False)
result, params = HEA.fit.fit.launch_fit(model_S, data, extended=False)
```

```python
assert HEA.fit.fit.check_fit(result)
```

### Plot the result

```python
# fig, ax, _ = plot_hist_fit_auto(df, branch, 
#                              models = [doubleCB], 
#                              models_names = ['\n $D^+ \\to K \\pi \\pi$'], 
#                              colors = ['g'], n_bins=n_bins,
#                              obs=obs, data_name=MC_name,
#                              params=params, latex_params=latex_params, 
#                              fontsize_leg=15., save_fig=False
#                             )

n_bins = int(round(sqrt(n_events)))
fig, ax, _ = plot_hist_fit_auto(df, branch, 
                             models = [model_S, crystalBallL_S, crystalBallR_S], 
                             models_names = ['\n $D^+ \\to K \\pi \\pi$', None, None], 
                             colors = 'r', n_bins=n_bins, models_types='mss',
                             obs=obs, data_name=MC_name, 
                             params=params, latex_params=latex_params, 
                             fontsize_leg=15., save_fig=False
                            )


ax.axvline(m_Dplus_PDG, color='b')
ax.text((m_Dplus_PDG - low_MC) / (high_MC - low_MC), 1.01, '$m(D^+)_{PDG}$',
        verticalalignment='bottom', horizontalalignment='center',
        transform=ax.transAxes,
        color='b', fontsize=20)

save_fig(fig, 'm_Kpipi_fit', folder_name=MC_name)
```
