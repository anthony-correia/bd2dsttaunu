## Get the $\chi^2$ distance between two the sWeighted LHCb data and reweighted or not MC
We define:
$$\chi^2 = \sum_{\text{bin i}} \frac{(\#1[i] - \#2[i])^2}{\#1[i] + \#2[i]}$$

In our case:
$$\chi^2 = \sum_{\text{bin i}} \frac{(\text{#MC}[i] - \text{#data}[i])^2}{\Delta(\text{#MC}[i]) + \Delta(\text{#data}[i])}$$

````python
true_data_weights = df['data']['sWeight']

chi2_result = pd.DataFrame()

d = {}

for column in plot_columns:
    if column != 'm_Kpipi' and\
        column != 'theta_X_reco' and\
        column != 'theta_L_reco' and\
        column != 'coschi_reco':
        
        chi2_dict = {}
        
        # low & high
        if column in column_ranges:
            low, high = column_ranges[column]
        else:
            low = None
            high = None
         
        # MC and LHCb data dataframes
        true_data = df['data'][column]
        original_MC = df['MC'][column]
        
        # chi2 when to weights applied to MC        
        chi2_dict['original'] = get_chi2_2samp(
            original_MC, true_data,
            n_bins=n_bins, low=low, high=high,
            weights1=None, weights2=true_data_weights
        )
        
        # With reweighting
        for reweight_name, df_MC in df['MC_weights'].items():
            MC_weights = df_MC['weight']
            
            chi2_dict[reweight_name] = \
                get_chi2_2samp(original_MC, true_data, 
                               n_bins=n_bins,
                               low=low, high=high,
                               weights1=MC_weights, weights2=true_data_weights)
        
        d[column] = pd.Series(chi2_dict.values(), index=chi2_dict.keys())
        
            
result_chi2 = pd.DataFrame(d)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(result_chi2)
        
````

## Compare other reweighting

````python
for column in plot_columns:
    if column in column_ranges:
        low, high = column_ranges[column]
    else:
        low = None
        high = None
    
    chi2 = {}
    chi2['original'] = str_number_into_latex(f"{result_chi2[column]['original']:.2g}")
    
    fig, ax = plot_divide_auto({'MC': df['MC'], 'sWeighted LHCb data':df['data']}, 
                               column, fig_name=column, label=f"Original MC, $\\chi^2 = {chi2['original']}$",
                               low=low, high=high, n_bins=n_bins, color='r',
                               weights=[None, df['data']['sWeight']],
                               pos_text_LHC={'ha': 'left'}, save_fig=False)

    colors = ['g', 'orange', 'grey', 'y', 'magenta', 'brown']
    for num, (name, df_MC) in enumerate(df['MC_weights'].items()):
        
        chi2[name] = str_number_into_latex(f"{result_chi2[column][name]:.2g}")
        
        plot_divide_auto({'MC': df['MC'], 'sWeighted LHCb data':df['data']}, 
                         column, 
                         label=name.replace('_', ' ')+f"$, \\chi^2 = {chi2[name]}$",
                         color=colors[num],
                         weights=[df_MC['weight'], df['data']['sWeight']],
                         ax=ax, folder_name=reweighted_MC_name,
                         low=low, high=high, n_bins=n_bins)

    ax.legend(fontsize=20, ncol=1)
````
